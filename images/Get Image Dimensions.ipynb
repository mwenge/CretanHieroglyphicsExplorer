```python
import cv2
import os
import subprocess

inputdir = "." 
for f in os.listdir(inputdir):          
    if not f.endswith("jpg"):
        continue                        
    input_file = inputdir + os.sep + f
    if os.path.isdir(input_file):       
        continue                        
    im = cv2.imread(input_file)
    h, w, c = im.shape
    if h / w <= 1.3:
        continue
    print(input_file, h, w)
    subprocess.check_call(['jpegtran', '-rotate', '90', '-outfile', input_file, input_file], stdout=subprocess.DEVNULL)
```

```python
import cv2
import os
import subprocess

id = "312"
inputdir = "." 
for f in os.listdir(inputdir):          
    if not f.startswith(id):
        continue                        
    input_file = inputdir + os.sep + f
    if os.path.isdir(input_file):       
        continue                        
    subprocess.check_call(['jpegtran', '-rotate', '90', '-outfile', input_file, input_file], stdout=subprocess.DEVNULL)
```

```python
import string

ids = [("000" + str(x))[-3:] for x in range(1, 400)]

for id in ids:
    for finfo in [('Facsimile', '-t'), ('Photo', '')]:
        suffix = finfo[0]
        x = finfo[1]
        cmd = ['convert']
        for e in string.ascii_lowercase[:6]:
            f = id + e + x + ".jpg"
            input_file = inputdir + os.sep + f
            if not os.path.isfile(input_file):
                continue
            if len(cmd) > 1:
                cmd.extend([f, "-append"])
            else:
                cmd.extend([f])
        if len(cmd) == 1:
            continue
        cmd.append(id + '-' + suffix + '.jpg')
        subprocess.check_call(cmd, stdout=subprocess.DEVNULL)
```
